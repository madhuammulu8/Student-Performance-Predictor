```python
import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/student_data.csv")
```

```python
df.head()
```

```python
# Store the 'Passed' feature in a new variable and remove it from the dataset
outcomes = df['passed']
data = df.drop('passed', axis = 1)

# Show the new dataset with 'Survived' removed
display(data.head())
```

```python
len(data)
```

```python
data = data.dropna()
```

```python
len(data)
```

```python
data = data.apply(lambda col: pd.factorize(col, sort=True)[0])
data.head()
```

```python
import numpy as np

from sklearn import preprocessing
scaler = preprocessing.MinMaxScaler()

scaler.fit(data)
scaled = scaler.transform(data)
train = pd.DataFrame(scaled, columns=data.columns)
train.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(train, outcomes, test_size=0.2,shuffle=False)
print(y_train)
```

```python
from sklearn.tree import DecisionTreeClassifier
# TODO: Define the classifier, and fit it to the data
model = DecisionTreeClassifier(max_depth=7 ,min_samples_leaf=6 ,min_samples_split=4)
model.fit(X_train, y_train)
```

```python
x=model.predict(X_train)
```

```python
# print(x)
```

```python
# Making predictions
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)

# Calculate the accuracy
from sklearn.metrics import accuracy_score
train_accuracy = accuracy_score(y_train, y_train_pred)
test_accuracy = accuracy_score(y_test, y_test_pred)
print('DecisionTree training accuracy is', train_accuracy)
print('DecisionTree test accuracy is', test_accuracy)
```

```python
from sklearn import svm
# TODO: Define the classifier, and fit it to the data
model = svm.SVC(kernel='poly')
model.fit(X_train, y_train)
```

```python
# Making predictions
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)

# Calculate the accuracy
from sklearn.metrics import accuracy_score
train_accuracy = accuracy_score(y_train, y_train_pred)
test_accuracy = accuracy_score(y_test, y_test_pred)
print('SVC training accuracy is', train_accuracy)
print('SVC test accuracy is', test_accuracy)
```
